```python
import pandas as pd

# 1. Load the verified dataset
df = pd.read_csv('C:/Users/sudha/Downloads/Portfolio Project/Superstore_Sales_Dataset/train.csv')

# 2. Handle the 11 missing Postal Codes
# We use 0 as a professional placeholder for unknown geographic data
df['Postal Code'] = df['Postal Code'].fillna(0).astype(int)

# 3. Standardize Column Names (Professional SQL convention: No spaces)
df.columns = [col.replace(' ', '_').replace('-', '_') for col in df.columns]

# 4. Convert and Verify Dates (Ensures SQL doesn't reject the format)
# Your CSV uses DD/MM/YYYY
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], dayfirst=True)

# 5. Export the Final Cleaned File
df.to_csv('Superstore_Cleaned_Final.csv', index=False)

print("Verification Complete:")
print(f"- Missing Dates found: 0")
print(f"- Missing Postal Codes fixed: 11")
print("- File 'Superstore_Cleaned_Final.csv' is ready for SQL Import.")
```
